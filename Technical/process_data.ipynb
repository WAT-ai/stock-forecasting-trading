```python
import pandas as pd
import numpy as np
import yfinance as yf

import functions

from sklearn.preprocessing import MinMaxScaler
```

```python
symbol = yf.Ticker('YUM')
per = '10y'

stock_data_df = functions.combine_all(symbol, per)
stock_data_df.head()
```

```python
#only keep price (close), volume, volatitlity, rsi , bollinger bands, and sma
features = ['Close', 'SMA50', 'Lower Band', 'Upper Band', 'Volatility', 'RSI']
select_data = stock_data_df[features]
select_data.head()
```

```python
date_col = select_data.index.date
# Convert the 'DateTime' column to datetime objects
date_col
```

```python
#normalize data using sklearns MinMax Scaler
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(select_data)

scaled_df = pd.DataFrame(scaled_data, columns=select_data.columns)
scaled_df.head()
scaled_df.insert(0, 'Date', date_col)
```
